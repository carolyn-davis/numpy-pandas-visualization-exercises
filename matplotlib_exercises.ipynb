```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def formula(x):
    return ((x**2)-x)-2
      
    
def graph(x):  
    y = pd.Series([formula(i) for i in range(10)])  
    y.plot()
    plt.annotate('origin (0,0)', xy=(0, 0))
    plt.show() 
    return y

test = graph(5)
```

```python
#Combine the figures you created in the last step into one large figure with 4 subplots.

def formula(x):
    return ((x**2)-x)-2

def squ_root_formula(x):
    return np.sqrt(x)   

def cubed_formula(x):
    return (x**3)

def raise_formula(x):
    return (2**x)

def graph(fun, x):  
    y = pd.Series([fun(i) for i in range(10)])  
    y.plot()
    plt.annotate('origin (0,0)', xy=(0, 0))
    plt.show() 
    return y   

plot_1 = graph(formula, 5)
plot_2 = graph(squ_root_formula, 5)
plot_3 = graph(cubed_formula, 5)
plot_4 = graph(raise_formula, 5)


plt.subplot(2,2,1)
plt.plot(plot_1)
plt.title('First Equation')

plt.subplot(2,2,2)
plt.plot(plot_2)
plt.title('Second Equation')

plt.subplot(2,2,3)
plt.plot(plot_3)
plt.text(5,650,'Third Equation',
         horizontalalignment='center')


plt.subplot(2,2,4)
plt.plot(plot_3)
plt.text(5,650,'Fourth Equation',
         horizontalalignment='center')
```
